```python
""" 
Q1.
a) Create a variable named var1 that stores an array of numbers from 0 to 30,
inclusive. Print var1 and its shape. Hint : arange
b) Change var2 to a validly-shaped two-dimensional matrix and store it in a
new variable called var2. Print var2 and its shape. Hint: Use the reshape function
c) Create a third variable, var3 that reshapes it into a valid three-dimensional shape.
Print var3 and its shape.
d) Use two-dimensional array indexing to set the first value in the second row of var2
to -1. Now look at var1 and var3. Did they change? Explain what’s
going on. (Hint: does reshape return a view or a copy?)
e) Another thing that comes up a lot with array shapes is thinking about how to
aggregate over specific dimensions. Figure out how the NumPy sum function
works (and the axis argument in particular) and do the following:
(i) Sum var3 over its second dimension and print the result.
(ii) Sum var3 over its third dimension and print the result.
(iii) Sum var3 over both its first and third dimensions and print the result.
f) Write code to do the following:
(i) Slice out the second row of var2 and print it.
(ii) Slice out the last column of var2 using the -1 notation and print it.
(iii) Slice out the top right 2 × 2 submatrix of var2 and print it.
"""
import numpy as np

# a) Create a variable named var1 that stores an array of numbers from 0 to 30, inclusive.
var1 = np.arange(30)
print(var1)
print(var1.shape)

# b) Change var2 to a validly-shaped two-dimensional matrix and store it in a new variable called var2.
var2 = var1.reshape(6, 5)  # Reshaping to 7 rows and 5 columns
print(var2)
print(var2.shape)

# c) Create a third variable, var3 that reshapes it into a valid three-dimensional shape.
var3 = var1.reshape(3, 5, 2)  # Reshaping to 3 layers, each with 5 rows and 2 columns
print(var3)
print(var3.shape)

# d) Use two-dimensional array indexing to set the first value in the second row of var2 to -1.
var2[1, 0] = -1
print("Updated var2:\n", var2)
print("var1 after updating var2:\n", var1)
print("var3 after updating var2:\n", var3)

# Explanation
"""
The changes made to var2 do not affect var1 or var3 because the reshape function returns a view of the original array only if possible. 
In this case, since var2 was created from var1, modifying it does not change the original data in var1, nor does it affect var3.
"""

# e) Using NumPy sum function with axis argument
# (i) Sum var3 over its second dimension and print the result.
sum_var3_axis1 = np.sum(var3, axis=1)
print("Sum of var3 over its second dimension:\n", sum_var3_axis1)

# (ii) Sum var3 over its third dimension and print the result.
sum_var3_axis2 = np.sum(var3, axis=2)
print("Sum of var3 over its third dimension:\n", sum_var3_axis2)

# (iii) Sum var3 over both its first and third dimensions and print the result.
sum_var3_axis0_and_2 = np.sum(var3, axis=(0, 2))
print("Sum of var3 over both first and third dimensions:\n", sum_var3_axis0_and_2)

# f) Slicing operations
# (i) Slice out the second row of var2 and print it.
second_row_var2 = var2[1]
print("Second row of var2:\n", second_row_var2)

# (ii) Slice out the last column of var2 using the -1 notation and print it.
last_column_var2 = var2[:, -1]
print("Last column of var2:\n", last_column_var2)

# (iii) Slice out the top right 2 × 2 submatrix of var2 and print it.
top_right_submatrix = var2[:2, -2:]
print("Top right 2x2 submatrix of var2:\n", top_right_submatrix)
```
